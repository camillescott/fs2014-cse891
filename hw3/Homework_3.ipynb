## Comments

There are a couple of observations to make in regards to performance. Firstly, counting sort is $O(N^2)$, while qsort is $O(NlogN)$. Thus, qsort will necessarily scale better with increasing $N$, and there will exist some value of $N$ for which qsort performs better than counting sort no matter how many processors are used. However, we also have to take into account memory bandwidth. We see that for counting sort on the Xeon Phi, scaling drops off after 64 threads; this is presumably because of the many IO operations hitting the limit of the Phi's memory throughput. Quicksort (qsort) also has much better locality in the way it subdivides an array, while counting sort is constantly missing the cache.

This case also illustrates why one should avoid using a parallel algorithm if there is a serial algorithm in a better complexity class.
